```python
#Relationship between Max Temperature and Relative Humidity
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ------------------------------------------------------------
# Load your dataset
# ------------------------------------------------------------
df = pd.read_excel(r"C:\Users\rakib\Documents\bangladesh_2000_2024.xlsx")

# Make sure the 'time' column is datetime type
df['time'] = pd.to_datetime(df['time'])
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month


# ------------------------------------------------------------
# 2️⃣ Scatter Plot: Max Temp vs Relative Humidity (colored by Month)
# ------------------------------------------------------------
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df, x='relative_humidity', y='temp_max', hue='month',
    palette='Spectral', alpha=0.6, edgecolor=None
)
plt.title("Relationship between Max Temperature and Relative Humidity (2000–2024)", fontsize=14)
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Max Temperature (°C)")
plt.legend(title="Month", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


```

```python
#Relationship between Max Temperature and Max Dewpoint
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# ------------------------------------------------------------
# Load dataset
# ------------------------------------------------------------
df = pd.read_excel(r"C:\Users\rakib\Documents\bangladesh_2000_2024.xlsx")

# Convert to datetime and extract year & month
df['time'] = pd.to_datetime(df['time'])
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month

# Drop rows with missing values (if any)
df = df.dropna(subset=['temp_max', 'dewpoint_max'])

# ------------------------------------------------------------
# 1️⃣ Overall Pearson Correlation
# ------------------------------------------------------------
r, p = pearsonr(df['temp_max'], df['dewpoint_max'])
print(f"Pearson Correlation: {r:.3f}, P-value: {p:.3e}")



# ------------------------------------------------------------
# 3️⃣ Scatter Plot: Temp vs Dewpoint (colored by Month)
# ------------------------------------------------------------
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df, x='dewpoint_max', y='temp_max', hue='month',
    palette='Spectral', alpha=0.6, edgecolor=None
)
plt.title("Relationship between Max Temperature and Max Dewpoint (2000–2024)", fontsize=14)
plt.xlabel("Max Dewpoint (°C)")
plt.ylabel("Max Temperature (°C)")
plt.legend(title="Month", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
#Relationship between Max Temperature and Max Wind Speed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# ------------------------------------------------------------
# Load dataset
# ------------------------------------------------------------
df = pd.read_excel(r"C:\Users\rakib\Documents\bangladesh_2000_2024.xlsx")

# Convert 'time' to datetime and extract year/month
df['time'] = pd.to_datetime(df['time'])
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month

# Drop missing values (if any)
df = df.dropna(subset=['temp_max', 'wind_max'])

# ------------------------------------------------------------
# 1️⃣ Compute Pearson Correlation
# ------------------------------------------------------------
r, p = pearsonr(df['temp_max'], df['wind_max'])
print(f"Pearson Correlation: {r:.3f}, P-value: {p:.3e}")



# ------------------------------------------------------------
# 3️⃣ Scatter Plot: Temp vs Wind (colored by Month)
# ------------------------------------------------------------
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df, x='wind_max', y='temp_max', hue='month',
    palette='Spectral', alpha=0.6, edgecolor=None
)
plt.title("Relationship between Max Temperature and Max Wind Speed (2000–2024)", fontsize=14)
plt.xlabel("Max Wind Speed (m/s)")
plt.ylabel("Max Temperature (°C)")
plt.legend(title="Month", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()


```
